```python
from transformers import pipeline

# Загружаем модель через pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Пример текста и вопроса
context = """
Deep learning is a class of machine learning methods that aims to learn data representations through artificial neural networks.
These methods are used in a variety of applications, including speech recognition, computer vision, and natural language processing.
"""
question = "What is deep learning used for?"

# Получаем ответ от модели
result = qa_pipeline(question=question, context=context)

# Выводим ответ
print("Answer:", result['answer'])
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
zero_shot_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Пример статьи
article = """
Your long article text goes here...
"""

# Генерация резюме
summary = summarizer(article, max_length=150, min_length=50, do_sample=False)
summary_text = summary[0]['summary_text']

# Анализ тональности
sentiment = sentiment_analyzer(summary_text)

# Классификация статьи
candidate_labels = ["sports", "politics", "economics"]
classification = zero_shot_classifier(summary_text, candidate_labels)

print("Summary:", summary_text)
print("Sentiment:", sentiment)
print("Classification:", classification['labels'][0])
```

```python
# Установка зависимостей
!pip install streamlit transformers pillow pyngrok -q

# Сохранение Streamlit приложения в файл
import streamlit as st
from transformers import pipeline
from PIL import Image

st.title("Анализ текста и изображений")

st.write("""
Это приложение анализирует текст и изображения:
1. Определяет категорию текста.
2. Генерирует описание загруженного изображения.
""")

@st.cache_resource
def load_models():
    # Загрузка моделей
    text_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
    image_captioning = pipeline("image-to-text", model="nlpconnect/vit-gpt2-image-captioning")
    return text_classifier, image_captioning

# Загрузка моделей
text_classifier, image_captioning = load_models()

# Категории текста
categories = ["спорт", "политика", "технологии", "культура", "экономика", "здравоохранение"]

# Поля ввода
text = st.text_area("Введите текст:")
uploaded_image = st.file_uploader("Загрузите изображение", type=["jpg", "png"])

# Кнопка анализа
if st.button("Анализировать"):
    if text.strip():
        # Классификация текста
        text_result = text_classifier(text, categories)
        st.subheader("Категория текста:")
        st.write(f"Категория: {text_result['labels'][0]} (уверенность: {text_result['scores'][0]:.2f})")

    if uploaded_image:
        # Описание изображения
        image = Image.open(uploaded_image)
        st.image(image, caption="Загруженное изображение", use_column_width=True)
        image_caption = image_captioning(image)
        st.subheader("Описание изображения:")
        st.write(image_caption[0]['generated_text'])

    if not text.strip() and not uploaded_image:
        st.warning("Пожалуйста, введите текст или загрузите изображение!")

# Боковая панель
st.sidebar.title("О приложении")
st.sidebar.info("""
Это приложение использует модели NLP и CV для анализа текста и изображений.
Вы можете вводить текст, загружать изображения, и приложение предоставит результат.
""")

# Запуск Streamlit-приложения
!streamlit run app.py &

# Установка и запуск ngrok для создания публичной ссылки
from pyngrok import ngrok

# Открываем туннель к локальному порту 8501

```
